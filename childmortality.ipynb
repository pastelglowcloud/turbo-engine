```python
import wbdata
import matplotlib as mpl
import matplotlib.pyplot as plt
import matplotlib.style as mplstyle
from matplotlib.pyplot import figure
import numpy as np
from cycler import cycler
data = np.random.randn(50)
plt.style.use('bmh')
```

```python
countries = ["IND","AFG","BGD","BTN","NPL","PAK","MDV","LKA"]
indicators = {'SH.DYN.MORT':'Child Mortality'}
```

```python
df = wbdata.get_dataframe(indicators, country=countries, convert_date=False)
dfu = df.unstack(level=0)
```

```python
plt.figure();dfu.plot();plt.title("Child Mortality in India and South Asia");plt.xlabel('Year');plt.ylabel('Child Mortality (Under-5) per 1000 live births');
plt.legend(bbox_to_anchor=(1, 1),
           bbox_transform=plt.gcf().transFigure)
plt.legend(prop={'size': 10})
figure(figsize=(40, 40), dpi=80)
plt.show()
```
